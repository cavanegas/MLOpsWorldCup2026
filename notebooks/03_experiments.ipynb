# 03 — Experimentos con MLflow

Barrido de 3 algoritmos (LogReg, GradientBoosting, XGBoost) con cross-validation
temporal. Cada corrida queda registrada en MLflow.

```python
import sys; sys.path.insert(0, '../src')
from worldcup2026.features.build_features import load_training_dataset
from worldcup2026.models.train import train_model

ds = load_training_dataset()
```

```python
for model_name, params in [
    ('logreg', {'C': 0.5}),
    ('logreg', {'C': 2.0}),
    ('gbt',    {'n_estimators': 250, 'max_depth': 3, 'learning_rate': 0.05, 'random_state': 42}),
    ('gbt',    {'n_estimators': 500, 'max_depth': 4, 'learning_rate': 0.03, 'random_state': 42}),
    ('xgb',    {'n_estimators': 400, 'max_depth': 4, 'learning_rate': 0.05, 'random_state': 42}),
]:
    try:
        tm = train_model(ds, model_name=model_name, params=params, register=False)
        print(model_name, params, '->', tm.metrics)
    except RuntimeError as e:
        print('skipped', model_name, ':', e)
```

Abre `mlflow ui --backend-store-uri ../mlruns` para comparar runs.
